```python
from pyspark import SparkConf,SparkContext
from pyspark.sql import SQLContext

import random
conf=SparkConf().setMaster("local").setAppName("SparkContext_Case2")
sc=SparkContext.getOrCreate(conf=conf)
print(sc)
sql_context=SQLContext(sc)
print(sql_context)
df=sql_context.read.csv('/opt/sparktest/SwimmerAnalysis.csv',header=True,inferSchema=True)
print("Show eight roles of table")
df.show(8)

print("Selecting swimTimeInSecond")
swimmerDF=df.select("swimTimeInSecond")
swimmerDF.show(6)

print("Selecting id and swimmerSpeed Column")
swimmerDF1=df.select("id","swimmerSpeed")
swimmerDF1.show(9)

print("Multiply swimmerSpeed column by 2")
swimmerDF2=df.select("id",df.swimmerSpeed*2)
swimmerDF2.show(5)

print("Selecting records where gender column is Male")
swimmerDF3=df.filter(df.Gender=='Male')
swimmerDF3.show()

print("Selecting records where gender is Male and occupation is Programmer")
swimmerDF4=df.filter((df.Gender=='Male')&(df.Occupation=='Programmer'))
swimmerDF4.show()

print("Selecting records where Occupation is Programmer and swimmerSpeed > 1.271")
swimmerDF5=df.filter((df.Occupation=='Programmer')&(df.swimmerSpeed>1.271))
swimmerDF5.show()


#df.printSchema()
#df.limit(5).toPandas()
sc.stop()
```
