```python
from pyspark import SparkConf,SparkContext
from pyspark.sql import SQLContext
from pyspark.sql.functions import *
import random
conf=SparkConf().setMaster("local").setAppName("SparkContext_Case2")
sc=SparkContext.getOrCreate(conf=conf)
print(sc)
sql_context=SQLContext(sc)
print(sql_context)
print("Read CSV Data From File")
corrData=spark.read.json(path='/opt/sparktest/testData.json')
corrData.show(6)

print("Calculate Mean of Each Column")
meanVal1=corrData.agg({"iv1":"avg","iv2":"avg","iv3":"avg"})
meanVal1.show()

print("Calculate Sample Variance of Each Column")
variance=corrData.agg({"iv1":"var_samp","iv2":"var_samp","iv3":"var_samp"})
variance.show()

print("Calculate Population Variance of Each Column")
variance=corrData.agg({"iv1":"var_pop","iv2":"var_pop","iv3":"var_pop"})
variance.show()

print("Counting number of data points in Each Column")
variance=corrData.agg({"iv1":"count","iv2":"count","iv3":"count"})
variance.show()

print("Calculating Summation,Mean,Standard Deviation on the First Column ")
sumAvgStdDev=corrData.agg(sum("iv1"),avg("iv1"),stddev_samp("iv1"))
sumAvgStdDev.show()

print("Calculating variance of first column,Mean of second column and standard deviation of third column ")
varAvgStdDev=corrData.agg({"iv1":"var_samp","iv2":"avg","iv3":"stddev_samp"})
varAvgStdDev.show()

print("Calculating Covariance between variables iv1 and iv2, iv2 and iv3, iv1 and iv3")
cov1=corrData.cov('iv1','iv2')
print("Covariance between v1 and v2:: ")
print(cov1)

cov2=corrData.cov('iv2','iv3')
print("Covariance between v2 and v3:: ")
print(cov2)

cov3=corrData.cov('iv1','iv3')
print("Covariance between v1 and v3:: ")
print(cov3)

print("Calculating Corellation between variables iv1 and iv2, iv2 and iv3, iv1 and iv3")
corr1=corrData.corr('iv1','iv2')
print("Corellation between v1 and v2:: ")
print(corr1)

corr2=corrData.corr('iv2','iv3')
print("Corellation between v2 and v3:: ")
print(corr2)

corr3=corrData.corr('iv1','iv3')
print("Corellation between v1 and v3:: ")
print(corr3)

print("Apply describe() function on each column")
dataDescription=corrData.describe()
dataDescription.show()

print("Applying describe() function on columns 'iv1' and 'iv2'")
dataDescriptioniv=corrData.describe(['iv1','iv2'])
dataDescriptioniv.show()

print("Applying summary() function on each columns")
dataDescriptioncolumns=corrData.summary()
dataDescriptioncolumns.show()

print("Applying summary() function with mean and on each columns")
dataDescriptioncolumns=corrData.summary(['mean','max'])
dataDescriptioncolumns.show()

sc.stop()
```
